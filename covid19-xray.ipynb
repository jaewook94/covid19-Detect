```python
import pandas as pd
import os
import shutil
```

```python
# Create the data for positive samples (GitHub)

FILE_PATH = "chestxray-dataset/metadata.csv"
IMAGES_PATH = "chestxray-dataset/images"
```

```python
df = pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head(30)
```

```python
TARGET_DIR = "Dataset/Covid"
# Making Covid folder in Dataset folder
if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("Covid folder created")
```

```python
cnt = 0

for (i,row) in df.iterrows():
    # we have to select PA view which means 'posteroanterior view'
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row["view"]=="PA":
        filename = row['filename']
        # [join] - Make image path binded by '/'
        image_path = os.path.join(IMAGES_PATH, filename)
        image_copy_path = os.path.join(TARGET_DIR, filename)
        # Copy with meta informations
        shutil.copy2(image_path, image_copy_path)
        # print("Moving Image",cnt)
        cnt +=1
```

```python
# Sampling of Images from Kaggle

import random
KAGGLE_FILE_PATH = "chest_xray_kaggle/train/NORMAL"
TARGET_NORMAL_DIR = "Dataset/Normal"
```

```python
image_names = os.listdir(KAGGLE_FILE_PATH)
```

```python
image_names
```

```python
# Randomlt shuffle the list of images to pick the first 196 images to create the dataset
random.shuffle(image_names)
```

```python
image_names
```

```python
for i in range(196):
    image_name = image_names[i]
    image_path = os.path.join(KAGGLE_FILE_PATH, image_name)
    
    target_path = os.path.join(TARGET_NORMAL_DIR, image_name)
    
    shutil.copy2(image_path,target_path)
    print("Copying image",i)
```
